```python
dataset = pd.read_excel("Food Consumption & Access Control Data (Admin) 2023-24 - All XYZ Location Offices-Mar 25.xlsb")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for day in dataset['Day'].unique():
    plt.figure(figsize=(8, 4))
    sns.barplot(x='Facility', y='Lunch Ordered previous day', 
                data=dataset[dataset['Day'] == day])
    plt.xticks(rotation=45)
    plt.title(f'Lunch Ordered on {day}')
    plt.xlabel('Facility')
    plt.ylabel('Lunch Ordered')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for day in dataset['Day'].unique():
    plt.figure(figsize=(8, 4))
    sns.barplot(x='Facility', y='Access Control Data (Footfall)', 
                data=dataset[dataset['Day'] == day])
    plt.xticks(rotation=45)
    plt.title(f'FootFall on {day}')
    plt.xlabel('Facility')
    plt.ylabel('Footfall')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.lineplot(x=dataset["Date"], y=dataset["Access Control Data (Footfall)"], label="Footfall", marker="o")
sns.lineplot(x=dataset["Date"], y=dataset["Lunch Ordered previous day"], label="Lunch Ordered", marker="o")
plt.xticks(rotation=45)
plt.title("Footfall vs. Lunch Ordered Over Time")
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.histplot(dataset['Diffrence =(H-G)'], bins=100, kde=True)
plt.axvline(0, color='red', linestyle="--", label="Perfect Match")
plt.title("Distribution of Lunch Order Differences")
plt.legend()
plt.show()
```

```python
facilities = dataset["Facility"].unique()

# Set figure size
plt.figure(figsize=(12, 6 * len(facilities)))

# Loop through each facility and plot separately
for i, facility in enumerate(facilities, 1):
    facility_dataset = dataset[dataset["Facility"] == facility]
    
    plt.subplot(len(facilities), 1, i)
    sns.lineplot(x=facility_dataset["Date"], y=facility_dataset["Diffrence =(H-G)"], marker='o', label=facility)
    
    plt.axhline(y=0, color='r', linestyle='--', label="No Difference")
    plt.xlabel("Date")
    plt.ylabel("Lunch Order Difference")
    plt.title(f"Lunch Order Difference Trend for {facility}")
    plt.xticks(rotation=45)
    plt.legend()

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
facilities = dataset["Facility"].unique()

# Set figure size
plt.figure(figsize=(12, 6 * len(facilities)))


for i, facility in enumerate(facilities, 1):
    facility_dataset = dataset[dataset["Facility"] == facility]

    plt.subplot(len(facilities), 1, i)
    sns.histplot(facility_dataset["Diffrence =(H-G)"], bins=100, kde=True, color="blue")

    plt.axvline(x=0, color='r', linestyle='--', label="Perfect Match")
    plt.xlabel("Lunch Order Difference")
    plt.ylabel("Count")
    plt.title(f"Distribution of Lunch Order Difference for {facility}")
    plt.legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
